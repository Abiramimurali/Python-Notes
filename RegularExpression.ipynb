```python
import re
string="ABCD4COMPUTER 765 PYTHON 687 abi"
pattern='COMPUTER'#extracting specific word
match=re.findall(pattern,string)
print(match)
```

```python
pattern=r'[0-9]+'
match=re.findall(pattern,string)
print(match)

pattern=r'[A-Za-z0-9]+'
match=re.findall(pattern,string)
print(match)

pattern=r'[^ ]+'  
match=re.findall(pattern,string)
print(match)

pattern=r'[^A ]+'  
match=re.findall(pattern,string)
print(match)

pattern=r'[^A]+'  
match=re.findall(pattern,string)
print(match)

pattern=r'[^P ]+'  
match=re.findall(pattern,string)
print(match)

pattern=r'[^P]+'  
match=re.findall(pattern,string)
print(match)

```
